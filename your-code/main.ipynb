```python
# your code here
 
# MEDIAN STEPS BY MONTH_ENCODED
steps_by_month = db_fit_cpy.groupby('Months_encoded').agg({'Steps':'median'}).reset_index()


plt.figure(figsize=(12, 4))

plt.subplot(1,2,1)
plt.bar(steps_by_month['Months_encoded'], steps_by_month['Steps'], color='skyblue', edgecolor='black')



# MEDIAN STEPS BY WORK_OR_WEEKEND
steps_by_week = db_fit_cpy.groupby('Work_or_Weekend').agg({'Steps':'median'}).reset_index()


plt.subplot(1,2,2)
plt.bar(steps_by_week['Work_or_Weekend'], steps_by_week['Steps'], color='skyblue', edgecolor='black')


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('figures.png')
plt.tight_layout()
plt.show()



# your code here
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

Acti = [db_fit_cpy['Minutes Lightly Active'],db_fit_cpy['Minutes Very Active'], db_fit_cpy['Minutes Sedentary']]
steps = [db_fit_cpy['Steps'],db_fit_cpy['Steps'],db_fit_cpy['Steps']]


for i in range(len(Acti)):
    plt.scatter(x=steps[i], y=Acti[i], alpha=0.70,s=70,color='skyblue', edgecolors='black')
    plt.xlabel('Activity',color='blue',style='oblique')
    plt.ylabel('Steps',color='blue',style='oblique')
    plt.title('Activity by steps')
    plt.show()
```
